### ***Python RegEx (Regular Expression)***   
***RegEx ব্যবহার করা হয় string এর মধ্যে pattern খোঁজা, মিলানো, replace করা, validate করা ইত্যাদির জন্য।***   
***Python-এ RegEx ব্যবহার করতে হয় re module দিয়ে।***   

***For Learning RegEX: https://www.w3schools.com/python/python_regex.asp***   
***Creating RegEx: https://regex101.com/***

```python
import re
def remove_html_tags(text):
    pattern = re.compile(r'<.*?>')
    return pattern.sub('', text)
```

***1️⃣ Python-এর RegEx module (re) ব্যবহার করা হচ্ছে***  
***2️⃣ একটা function বানানো হয়েছে যেটা text input নেয়***    
***3️⃣ RegEx pattern দিয়ে সব HTML tag খোঁজা হয়***   
***4️⃣ পাওয়া tag গুলোকে খালি string দিয়ে replace করা হয়***   
***5️⃣ শেষে পরিষ্কার text return করা হয়***   

```python
remove_html_tags(df['review'][3])
```

```python
df['review'].apply(remove_html_tags)
```

```python
df['review'] = df['review'].apply(remove_html_tags)
df.head()
```
